```python
# Import csv file for ACS Population Density
acs_pop_density_path = pd.read_csv("data_files\los_angeles_data\LA_pop_density_median_age_poverty_rate.csv")

# Read the CSV file into a DataFrame
la_pop_density_df = pd.DataFrame(acs_pop_density_path)

la_pop_density_df

```
